```python
import cv2
from ultralytics import YOLO

# Load the pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")

# Open the video file
cap = cv2.VideoCapture("/content/3006972-hd_1920_1080_24fps.mp4")
assert cap.isOpened(), "Error reading video file"

# Get video properties
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))# Define points for a line or region of interest
line_points = [(20, 400), (1080, 400)]

# Specify classes to count
classes_to_count = [0, 2]

# Initialize counters
object_counts = {cls: 0 for cls in classes_to_count}

# Initialize the video writer to save the output video
video_writer = cv2.VideoWriter("object_counting_output.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Process video frames
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break

    if im0.shape[:2] != (h, w):
        im0 = cv2.resize(im0, (w, h))

    # Perform object tracking
    results = model.track(im0, persist=True, classes=classes_to_count)

    # Draw the line for counting
    im0 = cv2.line(im0, line_points[0], line_points[1], (0, 255, 0), 2)

    # Get tracked objects
    for result in results:
        for det in result.boxes:
            if int(det.cls) in classes_to_count:
                bbox = det.xyxy[0].cpu().numpy().astype(int)
                class_id = int(det.cls)

                # Draw the bounding box on the frame
                cv2.rectangle(im0, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (255, 0, 0), 2)
                cv2.putText(im0, f'Class: {class_id}', (bbox[0], bbox[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

                # Update counts
                object_counts[class_id] += 1

    # Draw total counts on the frame
    count_text = "Counts: " + ", ".join([f'Class {cls}: {count}' for cls, count in object_counts.items()])
    cv2.putText(im0, count_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Write the annotated frame to the output video
    video_writer.write(im0)

# Release the video capture and writer objects
cap.release()
video_writer.release()


# Calculate and print the total number of objects detected
total_objects = sum(object_counts.values())
print(f"The total number of objects detected: {total_objects}")
# Print the total number of unique objects detected
print(f"The model detected {len(detected_objects)} unique object(s): {', '.join(map(str, detected_objects))}")

# Close all OpenCV windows
cv2.destroyAllWindows()
```
